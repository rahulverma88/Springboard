## Columns/rows to drop from final data frame

We can now decide which columns are redundant in the final data frame, and should be removed

1) The fluid in gallons and barrels are both reported, so obviously one is redundant. We keep the fluid measured in barrels, and lose the ['fluid_gal','fluid_gal_per_ft'] columns

2) There are file numbers, which should have no bearing on the machine learning model (columns 'file_no')

3) The Current_Well_Name and Lease_Number do not seem to have value for a machine learning model.

4) There are a number of columns related to the spatial location of the well. While spatial location should be very important, these columns denote geographical rather than geological information. So, I dropped ['Township','Range','Section']

5) Other names are related to how the oil field was named. This might be important, but I have decided to keep all that information within latitude and longitude. So I dropped ['Lease_Name','Field_Name','Original_Well_Name','Original_Operator','Current_Operator']

6) I am not sure of the meaning of these columns, and have decided to drop them as they are categorical variables which could have some meaning of no importance to the machine learning model: ['Well_Status','Well_Status_Date','QQ','CTB']

7) The difference between top and bottom gives the lateral length. We do not need all three variables. I decided to drop top.


```python
cols_to_drop = ['file_no','Current_Well_Name','Lease_Number','Township','Well_Type','Well_Status_Date','file_number','fluid_gal','fluid_gal_per_ft','top','Footages','Lease_Name','Field_Name','QQ','Range','CTB','Section','Well_Status','Original_Well_Name','Original_Operator','Current_Operator']
final_data.drop(labels=cols_to_drop,axis=1,inplace=True)
```

## Feature engineering ideas:
1) Flag for maturity: Classify based on Produced_pools, and then bin them based on TD - this should indicate some measure of maturity. This should also somewhat capture engineering data like pressure, temperature, oil composition which are not available here.

2) Fracture intensity: max_treat_press should give idea on fracture intensity, for a given Produced_pool.

### Other attempts to reduce R^2/RMSE:
1) Take only horizontal wells

2) Center categorical variables

3) Put more estimators in the random forest

### Maturity flag

For guessing at maturity, it is not enough to bin by TD, but one must also group by Produced_pools, which gives a rough idea of maturity within a formation.

I have decided to create 10 different levels of maturity, corresponding to 10 different bins for the TD's of each produced pool. Therefore, this will result in the creation of a new categorical variable, 'Maturity', whose value ranges from 1-10, and is also dependent on formation, which is added as a prefix.

Some formations don't have enough data points. There will be only 1 maturity value for that formation.

```python
final_data.head()
```

The produced_pools column was deleted from final_data, and must be added back in to make the maturity flag.

```python
# creating a new dataframe with_maturity to hold the final_data with maturity flag
with_maturity = final_data
```

```python
prod_pools = well_indices[['api','Produced_Pools']]
prod_pools.head()
```

```python
with_maturity = with_maturity.merge(prod_pools,how='inner',on='api')
```

```python
with_maturity['Maturity'] = with_maturity['TD']
```

```python
dict_pools = dict(list(with_maturity.groupby('Produced_Pools').Maturity))
```

```python
length_dict = {key: len(value) for key, value in dict_pools.items()}
```

```python
dict_pools['BIRDBEAR'].index
```

```python
grouped_data = with_maturity.groupby('Produced_Pools')
```

```python
for pool in with_maturity.Produced_Pools.unique():
    cur_td = grouped_data.get_group(pool).TD
    if len(cur_td) < 10: # if the formation has less than 10 data points, use one bin for all of them
        n_bins = 1
    else:
        n_bins = 10
    cur_range = max(cur_td) - min(cur_td) + 1 # 1 is added to take care of cases where only one data point exists
    cur_min = min(cur_td)
    bin_size = cur_range/n_bins
    vals = grouped_data.get_group(pool).TD.map(lambda x: (x -cur_min)// bin_size + 1).astype('str')
    #vals = [pool + vals.astype('str')]
    with_maturity.loc[vals.index,'Maturity'] = vals
```

```python
with_maturity.Maturity.unique()
```

```python
with_maturity['Maturity'] = with_maturity['Maturity'] + '_' + with_maturity['Produced_Pools'] # add in formation names to the bins
```

```python
with_maturity['Maturity'].head()
```

```python
with_maturity['Maturity'].unique()
```

```python
with_maturity[with_maturity['Produced_Pools']=='BAKKEN']['TD'].describe()
```

```python
with_maturity['Maturity'] = with_maturity['Maturity'].astype('category')
```

```python
maturity_dummies = pd.get_dummies(with_maturity['Maturity'],prefix='mtrty',drop_first=True)
```

```python
maturity_dummies.head()
```

```python
scaled_maturity = np.concatenate((scaled_data,maturity_dummies.values),axis=1)
```

```python
maturity_train, maturity_test, y_mat_train, y_mat_test = train_test_split(scaled_maturity,y,test_size=0.3,random_state=42)
```

```python
rf_maturity = RandomForestRegressor(n_estimators=500,random_state=100)
rf_maturity.fit(maturity_train,y_mat_train)
y_mat_pred = rf_maturity.predict(maturity_test)
rf_maturity.score(maturity_test,y_mat_test)
```
