```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from os import listdir
```

```python
wy_oil_prod = pd.read_csv('./OilGasProduction/Wyoming/EnigmaProduction_withCounty.csv')
```

```python
wy_oil_prod.head()
```

```python
np.sort(wy_oil_prod.county.unique())
```

```python
wy_oil_prod.columns
```

```python
wy_oil_cols = [x for x in wy_oil_prod.columns if 'oil' in x]
```

```python
wy_oil_cols = ['year','county'] + wy_oil_cols 
```

```python
wy_oil_small = wy_oil_prod[wy_oil_cols]
```

```python
wy_oil_small.head()
```

```python
wy_oil_month = wy_oil_small.groupby(['year','county']).agg('sum')
```

```python
wy_oil_month.head()
```

```python
wy_oil_month.reset_index(inplace=True)
```

```python
wy_oil_month = pd.melt(wy_oil_month, id_vars = ['year','county'],var_name='month')
```

```python
month_dict = dict(zip(['jan_oil', 'feb_oil', 'mar_oil', 'apr_oil', 'may_oil',
       'jun_oil', 'jul_oil', 'aug_oil', 'sep_oil', 'oct_oil', 'nov_oil',
       'dec_oil'],range(1,13)))
```

```python
month_dict
```

```python
wy_oil_month.month.unique()
```

```python
wy_oil_month.month = wy_oil_month.month.map(month_dict)
```

```python
wy_oil_month.month.unique()
```

```python
wy_oil_month.head()
```

```python
wy_oil_month['Date'] = wy_oil_month['month'].astype(str) + '/' + wy_oil_month['year'].astype(str)
```

```python
wy_oil_month.head()
```

```python
wy_oil_month.Date = pd.to_datetime(wy_oil_month.Date)
```

```python
wy_oil_month.head()
```

```python
wy_oil_month.Date = wy_oil_month.Date.dt.strftime('%m/%Y')
```

```python
wy_oil_month.head()
```

```python
wy_oil_month['API_County_Code'] = ["49%03d" % x for x in wy_oil_month['county']]
```

```python
api_codes = pd.read_csv('./OilGasProduction/State_County_API.csv')
```

```python
api_codes['API_County_Codes'] = ["%02d%03d" % (x,y) for x,y in zip(api_codes['API_State_Code'], api_codes['County_Code'])] 
```

```python
api_county_dict = dict(zip(api_codes['API_County_Codes'],api_codes['County_Name']))
```

```python
wy_oil_month['County_Name'] = wy_oil_month['API_County_Code'].map(api_county_dict)
```

```python
wy_oil_month.head()
```

```python
wy_oil_month.County_Name.unique()
```

```python
wy_unemp = pd.read_csv('./UnemploymentData/wy_unemployment.csv')
```

```python
wy_unemp.head()
```

```python
wy_unemp_melted = pd.melt(wy_unemp,id_vars = ['Time'],var_name='County_Name',value_name='Unemployment_Rate')
```

```python
wy_format_county = dict()

for county in wy_oil_month.County_Name.unique():
    wy_format_county[county] = ''.join([s for s in wy_unemp_melted.County_Name.unique() if county.lower() in s.lower()])
```

```python
wy_oil_month.County_Name = wy_oil_month.County_Name.map(wy_format_county)
```

```python
wy_oil_month.head()
```

```python
wy_oil_month.County_Name.unique()
```

```python
wy_oil_month = wy_oil_month[['Date','County_Name','value']]
```

```python
wy_oil_month.columns = ['Date','County_Name','Oil_Production']
```

```python
wy_oil_month = wy_oil_month.pivot(index='Date',columns='County_Name',values='Oil_Production')
```

```python
wy_oil_month.head()
```

```python
wy_oil_month = wy_oil_month.fillna(0)
```

```python
wy_oil_month.to_csv('./OilGasProduction/Wyoming/WYOilProdCounty.csv')
```
